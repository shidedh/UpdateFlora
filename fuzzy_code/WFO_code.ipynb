```python
def get_taxon_name(row):
    name = ""
    mouterde_genus = row['mouterde_genus']
    mouterde_epithet = row['mouterde_epithet']
    mouterde_infra = row['mouterde_infra']
    if mouterde_genus != "":
        name += f"{mouterde_genus}"
    if mouterde_epithet != "":
        name += f" {mouterde_epithet}"
    if mouterde_infra != "":
        name += f" {mouterde_infra}"
    return name
```

```python
vol1_df['mouterde_name'] = vol1_df.apply(get_taxon_name,axis=1)
vol2_df['mouterde_name'] = vol2_df.apply(get_taxon_name,axis=1)
vol3_df['mouterde_name'] = vol3_df.apply(get_taxon_name,axis=1)
```
